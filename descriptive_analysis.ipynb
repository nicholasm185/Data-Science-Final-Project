```python
plt.hist(cleaned["Price (Euros)"], range=(0,cleaned['Price (Euros)'].max()), bins=12)
plt.title("Frequency of Laptop Prices\n(range of 500 per bar)")
plt.xlabel("Laptop Price (Euros)")
plt.ylabel("Frequency")
plt.show()
```

```python
to_plot = [
    cleaned.loc[cleaned['Category'] == 'Ultrabook']['Price (Euros)'],
    cleaned.loc[cleaned['Category'] == 'Notebook']['Price (Euros)'],
    cleaned.loc[cleaned['Category'] == 'Netbook']['Price (Euros)'],
    cleaned.loc[cleaned['Category'] == 'Gaming']['Price (Euros)'],
    cleaned.loc[cleaned['Category'] == '2 in 1 Convertible']['Price (Euros)'],
    cleaned.loc[cleaned['Category'] == 'Workstation']['Price (Euros)']
    ]

plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')

plt.boxplot(to_plot, labels=['Ultrabook', 'Notebook', 'Netbook', 'Gaming', '2 in 1 Convertible', 'Workstation'])
plt.title("Price Ranges for Each Laptop Category")
plt.xlabel("Laptop Category")
plt.ylabel("Price (Euros)")
plt.show()
```
